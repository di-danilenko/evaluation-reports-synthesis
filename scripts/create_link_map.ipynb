```python
import pandas as pd
import os 
import glob

FOLDER = "/home/lt1412/Desktop/UNEG-repository/SDG-synthesis-coalition-tidy"
```

```python
# load the link map
df = pd.read_csv("/home/lt1412/Desktop/evaluation-reports-synthesis/link_map.csv")
display(df)
```

```python
# assign file path to each file based on document_id

def find_file_for_doc(doc_id, folder):
    """Find the lowest-numbered file for a given document_id."""
    matches = glob.glob(os.path.join(folder, f"{doc_id}_*.pdf"))  
    if not matches:
        return None
    # sort by the "index" number and return the lowest
    matches.sort(key=lambda x: int(x.rsplit('_', 1)[-1].split('.')[0]))
    return matches[0]

df['file_path'] = df['document_id'].apply(lambda doc_id: find_file_for_doc(doc_id, FOLDER))
```

```python
# drop rows where no file was found
df = df.dropna(subset=['file_path']).reset_index(drop=True)
```
